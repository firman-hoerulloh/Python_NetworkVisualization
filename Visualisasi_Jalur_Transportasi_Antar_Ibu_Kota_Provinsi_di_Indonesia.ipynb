```python
#IMPORT LIBRARY
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#CREATING THE GRAPH
indonesia = nx.Graph()
```

```python
#ADDING PROVINCIAL CAPITAL TO THE GRAPH
indonesia.add_nodes_from(["Banda Aceh",
    "Medan",
    "Padang",
    "Pekanbaru",
    "Tanjung Pinang",
    "Jambi",
    "Palembang",
    "Pangkal Pinang",
    "Bengkulu",
    "Bandar Lampung",
    "Jakarta",
    "Bandung",
    "Serang",
    "Semarang",
    "Yogyakarta",
    "Surabaya",
    "Denpasar",
    "Mataram",
    "Kupang",
    "Pontianak",
    "Palangka Raya",
    "Banjarmasin",
    "Samarinda",
    "Tanjung Selor",
    "Manado",
    "Gorontalo",
    "Palu",
    "Mamuju",
    "Makassar",
    "Kendari",
    "Ambon",
    "Sofifi",
    "Jayapura",
    "Manokwari",
    "Nabire",
    "Wamena",
    "Merauke",
    "Sorong"])
indonesia.nodes()

#SHOW THE GRAPH
nx.draw(indonesia, with_labels=True)
plt.show()
```

```python
#CREATING CONNECTIVITY BETWEEN NODES
indonesia.add_edge("Banda Aceh", "Medan")
indonesia.add_edge("Medan", "Padang")
indonesia.add_edge("Padang", "Pekanbaru")
indonesia.add_edge("Pekanbaru", "Jambi")
indonesia.add_edge("Jambi", "Palembang")
indonesia.add_edge("Palembang", "Bengkulu")
indonesia.add_edge("Palembang", "Bandar Lampung")
indonesia.add_edge("Palembang", "Pangkal Pinang")
indonesia.add_edge("Bengkulu", "Bandar Lampung")
indonesia.add_edge("Bandar Lampung", "Jakarta")
indonesia.add_edge("Jakarta", "Bandung")
indonesia.add_edge("Jakarta", "Semarang")
indonesia.add_edge("Jakarta", "Surabaya")
indonesia.add_edge("Jakarta", "Pontianak")
indonesia.add_edge("Jakarta", "Palangka Raya")
indonesia.add_edge("Jakarta", "Serang")
indonesia.add_edge("Jakarta", "Tanjung Pinang")
indonesia.add_edge("Jakarta", "Pangkal Pinang")
indonesia.add_edge("Bandung", "Semarang")
indonesia.add_edge("Bandung", "Yogyakarta")
indonesia.add_edge("Bandung", "Surabaya")
indonesia.add_edge("Bandung", "Serang")
indonesia.add_edge("Semarang", "Yogyakarta")
indonesia.add_edge("Semarang", "Surabaya")
indonesia.add_edge("Yogyakarta", "Surabaya")
indonesia.add_edge("Surabaya", "Denpasar")
indonesia.add_edge("Surabaya", "Mataram")
indonesia.add_edge("Surabaya", "Kupang")
indonesia.add_edge("Denpasar", "Mataram")
indonesia.add_edge("Kupang", "Mataram")
indonesia.add_edge("Pontianak", "Palangka Raya")
indonesia.add_edge("Palangka Raya", "Banjarmasin")
indonesia.add_edge("Banjarmasin", "Samarinda")
indonesia.add_edge("Samarinda", "Tanjung Selor")
indonesia.add_edge("Tanjung Selor", "Manado")
indonesia.add_edge("Manado", "Gorontalo")
indonesia.add_edge("Gorontalo", "Palu")
indonesia.add_edge("Palu", "Mamuju")
indonesia.add_edge("Mamuju", "Makassar")
indonesia.add_edge("Makassar", "Kendari")
indonesia.add_edge("Makassar", "Ambon")
indonesia.add_edge("Kendari", "Ambon")
indonesia.add_edge("Ambon", "Sofifi")
indonesia.add_edge("Sofifi", "Jayapura")
indonesia.add_edge("Jayapura", "Manokwari")
indonesia.add_edge("Manokwari", "Nabire")
indonesia.add_edge("Nabire", "Wamena")
indonesia.add_edge("Wamena", "Merauke")
indonesia.add_edge("Merauke", "Sorong")
indonesia.add_edge("Sorong", "Manokwari")

indonesia.edges()

#SHOW THE GRAPH
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(indonesia, k=0.3, seed=42)  # Adjust k for more spacing
nx.draw(indonesia, pos, with_labels=True, node_color='skyblue', font_size=10, font_color='black', edge_color='gray')
```

```python
#COUNTING DEGREE CENTRALITY
degree = nx.degree_centrality(indonesia)
degree
```

```python
#COUNTING BETWEENESS CENTRALITY
betweenness = nx.betweenness_centrality(indonesia)
betweenness
```

```python
#MOST IMPORTANT CITY
max_degree = max(degree, key=degree.get)
max_betweenness = max(betweenness, key=betweenness.get)

max_betweenness, max_betweenness_value = max_betweenness, betweenness[max_betweenness]
max_degree, max_degree_value = max_degree, degree[max_degree]

print("Kota terpenting berdasarkan betweenness:", max_betweenness, max_betweenness_value)
print("Kota terpenting berdasarkan degree:", max_degree, max_degree_value)
```

```python
#UPDATING THE NODES SIZE BASED ON DEGREE CENTRALITY
node_sizes = [10000 * degree[node] for node in indonesia.nodes()]

plt.figure(figsize=(12, 12))
pos = nx.spring_layout(indonesia, seed=42)  # For consistent layout
nx.draw(indonesia, pos, with_labels=True, node_size=node_sizes, node_color='skyblue', font_size=10, font_color='black', edge_color='gray')

plt.title("Konektivitas Ibu Kota setiap Provinsi dengan ukuran simpul berdasarkan Derajat Sentralitas")
plt.show()
```
